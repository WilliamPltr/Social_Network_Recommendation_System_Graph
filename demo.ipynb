# Demo: Exploring the Professional Social Graph

This notebook shows how to connect to the Neo4j database used by the FastAPI API,
run a few Cypher queries, and inspect example results.

> Note: make sure the docker stack is running (`docker compose up --build`)
> before executing the cells in this notebook.



```python
from neo4j import GraphDatabase

uri = "bolt://localhost:7687"
user = "neo4j"
password = "password"

driver = GraphDatabase.driver(uri, auth=(user, password))

with driver.session() as session:
    result = session.run("MATCH (u:User) RETURN u.id AS id LIMIT 5")
    print([record["id"] for record in result])
```
